```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# Define the updated MLP model architecture with an additional layer
class UpdatedMLP(nn.Module):
    def __init__(self):
        super(UpdatedMLP, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(784, 512),           # First Linear Layer
            nn.ReLU(),                     # ReLU Activation
            nn.Linear(512, 512),           # Second Linear Layer
            nn.BatchNorm1d(512),           # Batch Normalization
            nn.ReLU(),                     # ReLU Activation
            nn.Linear(512, 512),           # Third Linear Layer
            nn.BatchNorm1d(512),           # Batch Normalization
            nn.ReLU(),                     # ReLU Activation
            nn.Linear(512, 10)             # Output Layer
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)  # Flatten images from 28x28 to 784
        x = self.layers(x)
        return x

# Transformations for the MNIST dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Load MNIST dataset
train_dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root='./data', train=False, transform=transform, download=True)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Initialize the model, loss function, and optimizer
model = UpdatedMLP().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9)

# Lists to store accuracies and losses for plotting
train_accuracies, test_accuracies = [], []
train_losses, test_losses = [], []

# Training and evaluation function
def train_and_evaluate(model, num_epochs=20):
    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        correct_train = 0
        total_train = 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()

        train_loss = running_loss / len(train_loader)
        train_accuracy = 100 * correct_train / total_train
        train_losses.append(train_loss)
        train_accuracies.append(train_accuracy)

        # Evaluation phase
        model.eval()
        correct_test = 0
        total_test = 0
        running_test_loss = 0.0
        with torch.no_grad():
            for images, labels in test_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                running_test_loss += loss.item()

                _, predicted = torch.max(outputs, 1)
                total_test += labels.size(0)
                correct_test += (predicted == labels).sum().item()

        test_loss = running_test_loss / len(test_loader)
        test_accuracy = 100 * correct_test / total_test
        test_losses.append(test_loss)
        test_accuracies.append(test_accuracy)

        print(f"Epoch {epoch+1}/{num_epochs}, "
              f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.2f}%, "
              f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.2f}%")

# Train and evaluate the model
train_and_evaluate(model, num_epochs=20)

# Plotting Training and Test Accuracy
epochs = range(1, 21)

# Plot Training Accuracy
plt.figure(figsize=(8, 5))
plt.plot(epochs, train_accuracies, label='Train Accuracy', color='blue')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Training Accuracy per Epoch')
plt.legend()
plt.show()

# Plot Test Accuracy
plt.figure(figsize=(8, 5))
plt.plot(epochs, test_accuracies, label='Test Accuracy', color='green')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Test Accuracy per Epoch')
plt.legend()
plt.show()

# Plot Training Loss
plt.figure(figsize=(8, 5))
plt.plot(epochs, train_losses, label='Train Loss', color='red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss per Epoch')
plt.legend()
plt.show()

# Plot Test Loss
plt.figure(figsize=(8, 5))
plt.plot(epochs, test_losses, label='Test Loss', color='orange')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Test Loss per Epoch')
plt.legend()
plt.show()
```
